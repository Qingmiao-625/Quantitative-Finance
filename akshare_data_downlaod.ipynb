```python
df = ak.futures_zh_daily_sina(symbol="BU0")  # BU0 是沥青主力连续合约代码

end_date = pd.Timestamp('2025-06-26')  # 当前日期
start_date = end_date - pd.DateOffset(years=4)

df['date'] = pd.to_datetime(df['date'])
bu = df[(df['date'] >= start_date) & (df['date'] <= end_date)]

print(bu)
# 保存为 CSV 文件
bu.to_csv('data/bu.csv', index=False)
```
